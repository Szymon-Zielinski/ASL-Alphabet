# Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import cv2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D, MaxPooling2D, BatchNormalization
import keras_tuner as kt
```

# Geting data

```python
test_dir = 'asl_alphabet_test/asl_alphabet_test'
train_dir = 'asl_alphabet_train/asl_alphabet_train/'
```

```python
BATCH_SIZE = 5
IMG_SIZE = (224,224)
INPUT_SIZE = (224,224,3)
```

```python
def preprocess_image(img_path):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)  # Użyj preprocess_input zamiast normalizacji ręcznej, jeśli jesteś związany z konkretnym modelem
    return img_array


# def preprocess_image(img_tuple):
#     img_path, _ = img_tuple
#     img = image.load_img(img_path, target_size=IMG_SIZE)
#     img_array = image.img_to_array(img)
#     img_array = np.expand_dims(img_array, axis=0)
#     img_array = preprocess_input(img_array)
#     return img_array
```

```python
def random_images(classes):
    # if len(classes)<10:
    fig, ax = plt.subplots(len(classes), 6, figsize=(15,15))
    for i, label in enumerate(classes):
        ax[i,0].text(0.5,0.5,label,ha='center',va='center',fontsize=10)
        ax[i,0].axis('off')
        label_path = os.path.join(train_dir, label)
        list = os.listdir(label_path)
        for j in range(5):
            img_label = cv2.imread(os.path.join(label_path,list[j]))
            img_label = cv2.cvtColor(img_label, cv2.COLOR_BGR2RGB)
            ax[i,j+1].imshow(img_label)
            ax[i,j+1].axis('off')
    plt.tight_layout()
    plt.show()
```

```python
train_datagen = ImageDataGenerator(
    # preprocessing_function=preprocess_image,
    rescale=1.0/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.15
)
test_datagen = ImageDataGenerator(
   rescale=1/255.
)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)
validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)
```

```python
classes = list(train_generator.class_indices.keys())
```

```python
random_images(classes[:10])
```

# Hyperparameter tuning

```python
def build_model(hp):
    
    hp_activation_1 = hp.Choice('activation_1', values=['relu', 'tanh'])
    hp_layer_1 = hp.Int('layer_1', 16,1024,8)
    hp_layer_2 = hp.Int('layer_2', 16,1024,8)
    hp_layer_3 = hp.Int('layer_3', 16,1024,8)
    hp_layer_4 = hp.Int('layer_4', 16,1024,8)
    hp_layer_5 = hp.Int('layer_5', 16,1024,8)
    hp_layer_6 = hp.Int('layer_6', 16,1024,8)
    hp_layer_7 = hp.Int('layer_7', 16,1024,8)
    hp_layer_8 = hp.Int('layer_8', 16,1024,8)
    hp_drop_1 = hp.Float('dropout_1', 0,0.5,0.1)
    hp_drop_2 = hp.Float('dropout_2', 0,0.5,0.1)
    hp_drop_3 = hp.Float('dropout_3', 0,0.5,0.1)
    hp_drop_4 = hp.Float('dropout_4', 0,0.5,0.1)
    hp_learning_rate = hp.Choice('learning_rate', [1e-2,1e-3,1e-4])
    
    model = Sequential()
    
    model.add(Conv2D(hp_layer_1,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same',input_shape=INPUT_SIZE))
    model.add(Conv2D(hp_layer_2,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same'))
    model.add(BatchNormalization())
    model.add(MaxPooling2D(padding='same'))
    model.add(Dropout(hp_drop_1))

    # model.add(Conv2D(hp_layer_3,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same',input_shape=INPUT_SIZE))
    # model.add(Conv2D(hp_layer_4,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same'))
    # model.add(BatchNormalization())
    # model.add(MaxPooling2D(padding='same'))
    # model.add(Dropout(hp_drop_2))

    # model.add(Conv2D(hp_layer_5,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same',input_shape=INPUT_SIZE))
    # model.add(Conv2D(hp_layer_6,(3,3),activation=hp.Choice('activation', values=['relu', 'tanh']),padding='same'))
    # model.add(BatchNormalization())
    # model.add(MaxPooling2D(padding='same'))
    # model.add(Dropout(hp_drop_3))

    model.add(Flatten())
    model.add(Dense(hp_layer_7,activation=hp.Choice('activation', values=['relu', 'tanh'])))
    model.add(Dense(hp_layer_7,activation=hp.Choice('activation', values=['relu', 'tanh'])))
    model.add(Dropout(hp_drop_4))
    model.add(Dense(len(classes), activation='softmax'))

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate),
                                                    loss='categorical_crossentropy',
                                                    metrics=['accuracy'])
    return model
```

```python
tuner = kt.Hyperband(build_model,
                     objective='val_accuracy',
                     max_epochs=15,
                     factor=15,
                     directory='tuner',
                     project_name='sign')
```

```python
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
```

```python
train_img,train_labels = train_generator.next()
val_img,val_labels = validation_generator.next()
```

```python
tuner.search(train_img,train_labels,
            epochs=5,
            validation_data=(val_img,val_labels),
            callbacks=[early_stop])
```
